## üéØ OBJETIVO DA MISS√ÉO

Sua tarefa √© desenvolver um algoritmo em Python que:

    Identifique os tipos de armas (esp√©cies) envolvidas nas ocorr√™ncias

    Classifique as armas em categorias espec√≠ficas (calibres)

    Conte quantas armas de cada tipo foram apreendidas

## üîç DADOS DA OPERA√á√ÉO

O arquivo CSV cont√©m as seguintes colunas relevantes:

    ESPECIE_ARMA (descri√ß√£o do tipo de arma)

    CALIBRE_ARMA (calibre da arma)

    MARCA_ARMA (fabricante da arma)

```python
# Seu c√≥digo aqui:
import csv

exemplo = open('OCORRENCIAS_2025.csv')
leitor = csv.reader(exemplo, delimiter=';')

total_Espingarda = 0
total_Pistola = 0
total_Revolver = 0
total_Rifle = 0
total_Carabina = 0
total_Carabina_Fuzil = 0
total_Carabina_cartucheira = 0
total_Carabina_espingarda = 0
total_Fuzil = 0

total_Calibre_12 = 0
total_Calibre_22 = 0
total_Calibre_38 = 0
total_Calibre_9mm = 0
total_Outros_Calibres = 0

for linha in leitor:
    arma = linha[4].strip()  
    calibre = linha[5].strip() 

    if arma == "Espingarda":
        total_Espingarda += 1
    elif arma == "Pistola":
        total_Pistola += 1
    elif arma == "Revolver":
        total_Revolver += 1
    elif arma == "Rifle":
        total_Rifle += 1
    elif arma == "Carabina":
        total_Carabina += 1
    elif arma == "Carabina/Fuzil":
        total_Carabina_Fuzil += 1
    elif arma == "Carabina/cartucheira":
        total_Carabina_cartucheira += 1
    elif arma == "Carabina/espingarda":
        total_Carabina_espingarda += 1
    elif arma == "Fuzil":
        total_Fuzil += 1
    
    if calibre == "12":
        total_Calibre_12 += 1
    elif calibre == "22":
        total_Calibre_22 += 1
    elif calibre == "38":
        total_Calibre_38 += 1
    elif calibre == "9mm":
        total_Calibre_9mm += 1
    else:
        total_Outros_Calibres += 1

print(f"Total Espingarda: {total_Espingarda}")
print(f"Total Pistola: {total_Pistola}")
print(f"Total Revolver: {total_Revolver}")
print(f"Total Rifle: {total_Rifle}")
print(f"Total Carabina: {total_Carabina}")
print(f"Total Carabina/Fuzil: {total_Carabina_Fuzil}")
print(f"Total Carabina/cartucheira: {total_Carabina_cartucheira}")
print(f"Total Carabina/espingarda: {total_Carabina_espingarda}")
print(f"Total Fuzil: {total_Fuzil}")

print(f"Total Calibre 12: {total_Calibre_12}")
print(f"Total Calibre 22: {total_Calibre_22}")
print(f"Total Calibre 38: {total_Calibre_38}")
print(f"Total Calibre 9mm: {total_Calibre_9mm}")
print(f"Total Outros Calibres: {total_Outros_Calibres}")
```
